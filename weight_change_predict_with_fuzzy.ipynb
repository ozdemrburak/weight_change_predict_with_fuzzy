```python
#Çıktımızın aralığını veri setine göre belirledik. Diğer girdileri de aynı şekilde belirleyeceğim.
x_wchange = np.arange(df["Weight Change (lbs)"].min(), df["Weight Change (lbs)"].max(), 0.1) 
```

```python
x_duration = np.arange(1, 12, 1)
```

```python
x_cal = np.arange(0, 1950, 10)
```

```python
# Veri setinde object olarak girili ben sayısal olarak böyle sınıflandırdım. Aralıklandırma için de veri setinde 4'e bölünmüş.
x_pactivity = np.arange(1, 5, 1)
```

```python
x_age = np.arange(df["Age"].min(), df["Age"].max() + 1 , 1)
```

# Üyelik fonksiyonları tanımlama

```python
wage_low = fuzzy.trimf(x_age, [18, 30, 39])
wage_mid = fuzzy.trimf(x_age, [30, 44, 51])
wage_high = fuzzy.trimf(x_age, [44, 59, 59])

fig, (ax0) = plt.subplots(nrows = 1, figsize = (6, 3))

ax0.plot(x_age, wage_low, 'b', linewidth = 1, label = 'Düşük')
ax0.plot(x_age, wage_mid, 'g', linewidth = 1, label = 'Orta')
ax0.plot(x_age, wage_high, 'r', linewidth = 1, label = 'Yüksek')
ax0.legend()
plt.show()
```

```python
# Veri setinde fiziksel aktivite sedentary, lightly active, moderately active, very active olarak 4'e ayrılmış.
wpact_sdtry = fuzzy.trimf(x_pactivity, [1, 1, 2])      # Sedentary
wpact_liact = fuzzy.trimf(x_pactivity, [1, 2, 3])      # Lightly Active
wpact_mdact = fuzzy.trimf(x_pactivity, [2, 3, 4])      # Moderately Active
wpact_vact = fuzzy.trimf(x_pactivity, [3, 4, 4])       # Very Active

fig, (ax1) = plt.subplots(nrows = 1, figsize = (6, 3))

ax1.plot(x_pactivity, wpact_sdtry, 'b', linewidth = 1, label = 'Düşük')
ax1.plot(x_pactivity, wpact_liact, 'g', linewidth = 1, label = 'Biraz')
ax1.plot(x_pactivity, wpact_mdact, 'c', linewidth = 1, label = 'Orta')
ax1.plot(x_pactivity, wpact_vact, 'r', linewidth = 1, label = 'Yüksek')
ax1.legend()
plt.show()
```

```python
wcalory_normal = fuzzy.trimf(x_cal, [0, 250, 500])
wcalory_above_normal = fuzzy.trapmf(x_cal, [400, 500, 750, 800])
wcalory_gain = fuzzy.trimf(x_cal, [700, 875, 1000])
wcalory_extreme = fuzzy.trapmf(x_cal, [950, 1000, 1500, 2000])

fig, (ax2) = plt.subplots(nrows = 1, figsize = (6, 3))

ax2.plot(x_cal, wcalory_normal, 'b', linewidth = 1, label = 'Normal')
ax2.plot(x_cal, wcalory_above_normal, 'g', linewidth = 1, label = 'Normal Üstü')
ax2.plot(x_cal, wcalory_gain, 'c', linewidth = 1, label = 'Kilo alımı')
ax2.plot(x_cal, wcalory_extreme, 'r', linewidth = 1, label = 'Aşırı')
ax2.legend()
plt.show()
```

```python
wduration_short = fuzzy.trimf(x_duration, [0, 0, 4])
wduration_normal = fuzzy.trimf(x_duration, [3, 6, 8])
wduration_long = fuzzy.trimf(x_duration, [7, 12, 12])

fig, (ax4) = plt.subplots(nrows = 1, figsize=(6, 3))

ax4.plot(x_duration, wduration_short, 'b', linewidth= 1, label= 'Kısa')
ax4.plot(x_duration, wduration_normal, 'g', linewidth= 1, label= 'Normal')
ax4.plot(x_duration, wduration_long, 'r', linewidth= 1, label= 'Uzun')
ax4.set_title('Süre (Hafta)')
ax4.legend()
plt.show()
```

```python
# 1 lbs üstü aldıysa kilo almış sayaycağız, 1lbs kilo verdiyse kilo vermiş sayacağız !
wchange_loss = fuzzy.trimf(x_wchange, [-35, -15, 0])
wchange_stable = fuzzy.trimf(x_wchange, [-5, 0, 5])
wchange_gain = fuzzy.trimf(x_wchange, [0, 5, 35])

fig, (ax3) = plt.subplots(nrows=1, figsize=(6, 3))

ax3.plot(x_wchange, wchange_loss, 'b', linewidth=1.5, label='İyi')
ax3.plot(x_wchange, wchange_stable, 'g', linewidth=1.5, label='Düşük')
ax3.plot(x_wchange, wchange_gain, 'r', linewidth=1.5, label='Çok düşük')
ax3.set_title('Kilo Değişimi')
ax3.legend()
plt.show()
```

# Kuralların tanımlanması ve ctrl modülü
https://pythonhosted.org/scikit-fuzzy/api/skfuzzy.control.html

```python
from skfuzzy import control as ctrl
```

```python
calory = ctrl.Antecedent(x_cal, 'kalori')
pactivity = ctrl.Antecedent(x_pactivity, 'fiziksel_aktivite')
age = ctrl.Antecedent(x_age, 'yaş')
duration = ctrl.Antecedent(x_duration, 'süre')
wchange = ctrl.Consequent(x_wchange, 'kilo_degişimi')
```

```python
#Üyelik fonksiyonlarını eşleştirdik böylece kolayca kullanıcaz. 

calory['normal'] = wcalory_normal
calory['above_normal'] = wcalory_above_normal
calory['gain'] = wcalory_gain
calory['extreme']= wcalory_extreme

pactivity['sdtry'] = wpact_sdtry
pactivity['liact'] = wpact_liact
pactivity['mdact'] = wpact_mdact
pactivity['vact'] = wpact_vact

age['young'] = wage_low
age['mid'] = wage_mid
age['old'] = wage_high

duration['short'] = wduration_short
duration['normal'] = wduration_normal
duration['long'] = wduration_long

wchange['loss'] = wchange_loss
wchange['stable'] = wchange_stable
wchange['gain'] = wchange_gain
```

```python
#Kuralların tanımlanması
#bazı kuralları gereksiz yazdığımı farkettim, örn. 4 ve 8'den birini silebiliriz. Bunu da deneyeceğim.
rule1 = ctrl.Rule(calory['extreme'] & ~pactivity['vact'] & ~age["young"], wchange['gain'])
rule2 = ctrl.Rule(calory['extreme'] & pactivity['vact'] & ~age['old'], wchange['stable'])
rule3 = ctrl.Rule(calory['extreme'] & pactivity['mdact'], wchange['stable'])
#rule4 = ctrl.Rule(calory['normal'] & pactivity['sdtry'], wchange['stable'])
rule5 = ctrl.Rule(calory['extreme']  & ~pactivity['sdtry'] & ~age['old'], wchange['stable'])
rule6 = ctrl.Rule(calory['gain'] & ~(pactivity['mdact'] | pactivity['vact']) & ~duration['short'] , wchange['gain'])
rule7 = ctrl.Rule(calory['above_normal'] & ~(pactivity['mdact'] | pactivity['vact']), wchange['gain'])
rule8 = ctrl.Rule(calory['normal'] & ~(pactivity['mdact'] | pactivity['vact']), wchange['stable'])
rule9 = ctrl.Rule(calory['normal'] & pactivity['vact'], wchange['loss'])
rule10 = ctrl.Rule((pactivity['vact'] | pactivity['mdact']) & ~calory['extreme'] &(~age['old']) & ~duration['short'], wchange['loss'])
rule11 = ctrl.Rule((calory['gain'] | calory['extreme']) & age['old'] & ~duration['short'], wchange['gain'])
rule12 = ctrl.Rule(calory['extreme'] & pactivity['liact'] & ~age['young'], wchange['gain'])
```

```python
wchange_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule5, rule6, rule7, rule8, rule9,rule10, rule11, rule12])
```

```python
wchange_sim = ctrl.ControlSystemSimulation(wchange_ctrl)
```
